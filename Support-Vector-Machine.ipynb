```python
import pandas as pd
from sklearn import svm
import matplotlib.pyplot as plt
import numpy as np


def set_ax_gray(ax):
    ax.patch.set_facecolor("gray")
    ax.patch.set_alpha(0.1)
    ax.spines['right'].set_color('none')  # 设置隐藏坐标轴
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.grid(axis='y', linestyle='-.')

data = pd.read_csv('xigua_data3.0.csv')
x = data.iloc[:, [7]].values
y = data.iloc[:, 8].values

gamma = 10
C = 1

ax = plt.subplot()
set_ax_gray(ax)
ax.plot(np.linspace(0.0, 1.0, num=len(y)),y,label='data')

for gamma in [1, 10, 100, 1000]:
    print(f'gamma={gamma}')
    svr = svm.SVR(kernel='rbf', gamma=gamma)
    svr.fit(x, y)
    p_label = svr.predict(x)

    from sklearn.metrics import r2_score
    r2_score=r2_score(y, p_label)#计算r平方    
    
    print(f'r2_score={r2_score}')

    ax.plot(np.linspace(0.0, 1.0), svr.predict(np.linspace(0.2, 0.8).reshape(-1, 1)),
            label='gamma={}, C={}'.format(gamma, C))
    for i in range(len(x)):
        print("密度： {:.2f}, 预测含糖量： {:.2f}, 真实含糖量： {:.2f}".format(x[i][0], p_label[i], y[i]))
        
ax.legend(loc='upper left')
ax.set_xlabel('Density')
ax.set_ylabel('Sugar Content')

plt.show()
```

```python
import pandas as pd
from sklearn import svm
import matplotlib.pyplot as plt
import numpy as np
from libsvm import svmutil
from libsvm.svmutil import *



def set_ax_gray(ax):
    ax.patch.set_facecolor("gray")
    ax.patch.set_alpha(0.1)
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.grid(axis='y', linestyle='-.')



data = pd.read_csv('xigua_data3.0.csv')

X = data.iloc[:, [7]].values
y = data.iloc[:, 8].values

gamma = 10
C = 1


ax = plt.subplot()
set_ax_gray(ax)
ax.plot(np.linspace(0.0, 1.0, num=len(y)),y,label='data')

for gamma in [3, 30, 300, 3000]:
    prob = svm_problem(y, X)
    param = svm_parameter('-s 3 -t 2 -g {} -c {}'.format(gamma, C))#创建一个SVM参数对象param，设置SVM类型为C-SVC（-s 3）、核函数为RBF（-t 2）、gamma值为当前循环中的gamma值、惩罚系数为C
    m = svm_train(prob, param)#根据gamma值和prob对象进行训练
    p_label, p_acc, p_val = svm_predict(y, X, m)
    ax.plot(np.linspace(0.0, 1.0, num=len(p_label)), p_label,
            label='gamma={}, C={}'.format(gamma, C))#横坐标密度 纵坐标预测标签
    for i in range(len(X)):
        print("密度： {:.2f}, 预测含糖量： {:.2f}, 真实含糖量： {:.2f}".format(X[i][0], p_label[i], y[i]))

ax.legend(loc='upper left')
ax.set_xlabel('Density')
ax.set_ylabel('Sugar Content')

plt.show()
```
